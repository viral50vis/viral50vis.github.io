```python
base_dir = '../data/countries/'
week_mean_dir = '../data/week_mean/'

dataframe_collection = {}

for date in dates:
    dataframe_collection[date] = pd.DataFrame()

for directory in os.listdir(base_dir):
    if directory == ".DS_Store":
        continue
    print(directory)
    country_dir = base_dir + directory + "/" + directory + "_stats/" + directory + "_mean.csv"
    try:
        CC = pc.country_name_to_country_alpha3(directory, cn_name_format="default")
    except:
        CC = "GLO"
    temp = pd.read_csv(country_dir)
    temp.insert(1, "CC", CC)
    for index, row in temp.iterrows():
        dataframe_collection[row['date']] = dataframe_collection[row['date']].append(row)
```

```python
for date in dates:
    dataframe_collection[date] = dataframe_collection[date].drop(columns=['date'])
```

```python
for date in dates:
    dataframe_collection[date] = dataframe_collection[date].set_index('CC')
```

```python
comb_json = {}
for i, date in enumerate(dates):
    comb_json[date] = dataframe_collection[date].to_json(orient='index')
    comb_json[date] = json.loads(comb_json[date])

with open('../data/week_mean.json', 'w', encoding='utf-8') as f:
    json.dump(comb_json, f)
```

```python
for i, date in enumerate(dates):
    dataframe_collection[date].to_csv(week_mean_dir + str(i) + ".csv", index=True)
```
